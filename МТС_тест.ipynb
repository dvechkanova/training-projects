## Задание 2.1. На основании данных в папке «Отчеты» необходимо подключиться ко всем отчетам, построить сводную таблицу и настроить вывод метрик для каждой кампании

```python
#Создадим сводную таблицу по первому отчету
no_res_1 = (report_1['result'][report_1['result'].isin(['автоответчик', 'тишина'])].count() and 
     report_1['result'].isna().sum() or report_1['duration'][report_1['duration'] <= 3].count())
company_1 = pd.DataFrame()
company_1['База'] = report_1.groupby('Кампания')['msisdn'].nunique()

company_1 = company_1.reset_index()

company_1['Дозвон, %'] = round(100 - no_res_1/company_1['База']*100, 2)
company_1['Конверсия, %'] = round(report_1[report_1['result'] == 'согласие']['result'].count()/company_1['База']*100, 2)
company_1['start_day'] = report_1.call_start_time.min()
company_1['end_day'] = report_1.call_start_time.max()
company_1['Минуты'] = round(report_1.duration.sum()/60, 2)

company_1
```

```python
#Создадим сводную таблицу по второму отчету
no_res_2 = (report_2['result'][report_2['result'].isin(['автоответчик', 'тишина'])].count() and 
     report_2['result'].isna().sum() or report_2['duration'][report_2['duration'] <= 3].count())
company_2 = pd.DataFrame()
company_2['База'] = report_2.groupby('Кампания')['msisdn'].nunique()

company_2 = company_2.reset_index()

company_2['Дозвон, %'] = round(100 - no_res_2/company_2['База']*100, 2)
company_2['Конверсия, %'] = round(report_2[report_2['result'] == 'согласие']['result'].count()/company_2['База']*100, 2)
company_2['start_day'] = report_2.call_start_time.min()
company_2['end_day'] = report_2.call_start_time.max()
company_2['Минуты'] = round(report_2.duration.sum()/60, 2)

company_2
```

```python
#Создадим сводную таблицу по первому отчету
no_res_3 = (report_3['result'][report_3['result'].isin(['автоответчик', 'тишина'])].count() and 
     report_3['result'].isna().sum() or report_3['duration'][report_3['duration'] <= 3].count())
company_3 = pd.DataFrame()
company_3['База'] = report_3.groupby('Кампания')['msisdn'].nunique()

company_3 = company_3.reset_index()

company_3['Дозвон, %'] = round(100 - no_res_3/company_3['База']*100, 2)
company_3['Конверсия, %'] = round(report_3[report_3['result'] == 'согласие']['result'].count()/company_3['База']*100, 2)
company_3['start_day'] = report_3.call_start_time.min()
company_3['end_day'] = report_3.call_start_time.max()
company_3['Минуты'] = round(report_3.duration.sum()/60, 2)

company_3
```

```python
#Соеденим три сводные таблицы в один отчет
report = pd.concat([company_1,company_2, company_3], ignore_index= True)
# поправим вывод даты
report['start_day'] = pd.to_datetime(report['start_day']).dt.date
report['end_day'] = pd.to_datetime(report['end_day']).dt.date
report
```

## Задание 1.2. Построить сводную таблицу, где вывести call_status и result и настроить вывод метрик для каждой кампании

```python
# создадим сводную таблицу с результатами и статусом звонка по первому отчету 
status_res_1 = (report_1.groupby(['call_status', 'result'])
                .agg({'msisdn':'count', 'duration':'median'}))
status_res_1.columns = ['Отчет №1. Число элементов в столбце msisdn', 'Среднее время (медиана)']
status_res_1['Доля от количества уникальных номеров, %'] = round(status_res_1['Отчет №1. Число элементов в столбце msisdn']/
                                                            report_1['msisdn'].nunique()*100, 2)
status_res_1
```

```python
# создадим сводную таблицу с результатами и статусом звонка по второму отчету 
status_res_2 = (report_2.groupby(['call_status', 'result'])
                .agg({'msisdn':'count', 'duration':'median'}))
status_res_2.columns = ['Отчет №2. Число элементов в столбце msisdn', 'Среднее время (медиана)']
status_res_2['Доля от количества уникальных номеров, %'] = round(status_res_2['Отчет №2. Число элементов в столбце msisdn']/
                                                            report_2['msisdn'].nunique()*100, 2)
status_res_2
```

```python
# создадим сводную таблицу с результатами и статусом звонка по третьему отчету 
status_res_3 = (report_3.groupby(['call_status', 'result'])
                .agg({'msisdn':'count', 'duration':'median'}))
status_res_3.columns = ['Отчет №3. Число элементов в столбце msisdn', 'Среднее время (медиана)']
status_res_3['Доля от количества уникальных номеров, %'] = round(status_res_3['Отчет №3. Число элементов в столбце msisdn']/
                                                            report_3['msisdn'].nunique()*100, 2)
status_res_3
```

```python
#обобщим данные из трех отчетов в общей сводной таблице
status_res = (status_res_1.merge(status_res_2, on=['call_status', 'result'], how='outer')
              .merge(status_res_3, on=['call_status', 'result'], how='outer'))
status_res
```
